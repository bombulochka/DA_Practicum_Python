```python
def pop_platforms(a,b,c,d,e):
    plt.plot(games_gr[games_gr['platform']==a]['year_of_release'], games_gr[games_gr['platform']==a]['all_sales'],
            games_gr[games_gr['platform']==b]['year_of_release'], games_gr[games_gr['platform']==b]['all_sales'],
            games_gr[games_gr['platform']==c]['year_of_release'], games_gr[games_gr['platform']==c]['all_sales'],
            games_gr[games_gr['platform']==d]['year_of_release'], games_gr[games_gr['platform']==d]['all_sales'],
            games_gr[games_gr['platform']==e]['year_of_release'], games_gr[games_gr['platform']==e]['all_sales'])
    plt.grid(visible=True)
    plt.xlabel('–ì–æ–¥ –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä—ã')
    plt.ylabel('–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏')
    plt.legend([a,b,c,d,e])
    plt.title('–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º')
```

```python
pop_platforms(*pop)
```

#### 4. –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞

##### 4.1. –ü–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

```python
def portrait(df, region, param):
    print(df.groupby(param)[region].sum().sort_values(ascending=False)[:5])
```

```python
#–¢–û–ü-5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–∏
portrait(games, 'na_sales', 'platform')
```

```python
#–¢–û–ü-5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ï–≤—Ä–æ–ø—ã
portrait(games, 'eu_sales', 'platform')
```

```python
#–¢–û–ü-5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –Ø–ø–æ–Ω–∏–∏
portrait(games, 'jp_sales', 'platform')
```

```python
pop_platforms = games.pivot_table(index='platform', values=['na_sales', 'eu_sales', 'jp_sales'], aggfunc='sum')
pop_platforms
```

```python
pop_platforms.plot(kind='barh',stacked=True, title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –∏–≥—Ä–æ–≤—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º', xlabel='–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∏–≥—Ä, –º–ª–Ω');
```
