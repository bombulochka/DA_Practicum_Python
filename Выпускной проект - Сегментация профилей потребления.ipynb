# E-commerce — Выявление профилей потребления

# Задачи
Сегментировать покупателей по профилю потребления:
1. Провести исследовательский анализ данных

2. Сегментировать покупателей на основе истории их покупок
* На основе сегментации разработать персонализированные предложения для покупателей
* Проанализировать покупки и сегментировать товары на основе принадлежности к той или иной товарной
категории

3. Сформулировать и проверить статистические гипотезы

# Цели
1. Исходя из анализа сегментов покупателей, разработать рекомендации о том, каким клиентам и когда следует рассылать рекламные предложения о каких категориях товаров.
2. Разработать стратегию по увеличению количества продаваемого товара для каждого сегмента.
3. Предложить способы повышения возвращаемости покупателей для различных сегментов.

# Описание данных
* date — дата заказа;
* customer_id — идентификатор покупателя;
* order_id — идентификатор заказа;
* product — наименование товара;
* quantity — количество товара в заказе;
* price — цена товара

# 1. Предобработка данных

```python
# Загрузка библиотек
import pandas as pd
import numpy as np
from datetime import datetime as dt
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Загрузка данных
orders = pd.read_csv('https://code.s3.yandex.net/datasets/ecom_dataset_upd.csv')
orders.head(20)
```

```python
orders.info()
```

В таблице 7474 записи, пропуски отсутствуют. У столбца с датой некорректный формат.

```python
# Приведем столбец с датами к формату даты
orders['date'] = pd.to_datetime(orders['date'], format='%Y%m%d%H%M')
orders['date']
```

```python
# Проверим наличие явных дубликатов
orders.duplicated().sum()
```

Полные строки-дубликаты в данных отсутствуют.

```python
# Проверим дубликаты по клиенту, заказу и товару
orders[['customer_id', 'order_id', 'product']].duplicated().sum()
```

Обнаружено 1886 дубликатов по клиенту, заказу и товару.

```python
# Устраним дубликаты
orders = orders.drop_duplicates(subset=['customer_id', 'order_id', 'product'])
orders[['customer_id', 'order_id', 'product']].duplicated().sum()
```

В исходных данных было 7474 записи. Пропуски в данных отсутствуют. Столбец с датой приведен к формату даты. Устранено 1886 дубликатов по клиенту, заказу и товару.

```python
# Проверим наличие заказов, совершенных более чем одним покупателем
orders_grouped = orders.groupby('order_id', as_index=False)['customer_id'].nunique()
incorrect_orders = orders_grouped[orders_grouped['customer_id'] > 1]['order_id']
incorrect_orders.count()
```

```python
# Отфильтруем некорректные заказы
orders = orders[~orders['order_id'].isin(incorrect_orders)]
orders.shape[0]
```

# 2. Исследовательский анализ данных

## 2.1. Обработка аномальных значений

### 2.1.2. По сумме чека

```python
# Добавим столбец с выручкой
orders['revenue'] = orders['quantity'] * orders['price']
orders.head(10)
```

```python
# Посчитаем чек каждой покупки
revenue_by_order = orders.groupby('order_id',as_index=False)['revenue'].sum()
revenue_by_order.sort_values(by='revenue', ascending=False)
```

```python
revenue_by_order['revenue'].hist(bins=50)
plt.title('Выручка в разрезе заказов')
plt.xlabel('Чек покупки, р')
plt.ylabel('Количество покупок');
```

```python
# revenue_by_order[revenue_by_order < 10000].hist(bins=50)
revenue_by_order['revenue'].hist(range=(0, 10000), bins=50)
plt.title('Выручка в разрезе заказов')
plt.xlabel('Чек покупки, р')
plt.ylabel('Количество покупок');
```

По гистограммам видно, что подавляющая часть покупок не превышает 4 тыс.р. При этом присутствуют крупные заказы до 700 тыс.р.

```python
# Построим диаграмму рассеяния
x = pd.Series(range(0, len(revenue_by_order)))
plt.scatter(x, revenue_by_order['revenue'])
plt.grid(True)
plt.title('Количество заказов на покупателя')
plt.ylabel('Количество заказов');
```

Присутствуют выбросы, есть заказ на сумму 675 тыс.р.

```python
# Посмотрим 95-й и 99-й перцентили по чеку покупки
revenue_by_order = orders.groupby('order_id', as_index=False)['revenue'].sum()
np.percentile(revenue_by_order['revenue'], [95, 99])
```

Не более 5 процентов заказов дороже 3776 руб., и не более 1 процента заказов дороже 7679 руб.

<div class="alert alert-info" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid " ><b> Комментарий студента : </b>

Видим аномалию по диаграмме рассеяния - заказ на сумму 675 тыс.р.

</div>

```python
# Отфильтруем этот заказ
expensive_order = revenue_by_order[revenue_by_order['revenue'] == revenue_by_order['revenue'].max()]
orders = orders[~orders['order_id'].isin(expensive_order['order_id'])]
```

### 2.1.3. По количеству заказов на покупателя

```python
# Посчитаем количество заказов для каждого покупателя
orders_by_customer = orders.groupby('customer_id', as_index=False).agg({'order_id':'nunique'})
orders_by_customer.sort_values(by='order_id',ascending=False).head(10)
```

```python
# Построим график
x = pd.Series(range(0, len(orders_by_customer)))
plt.scatter(x, orders_by_customer['order_id'])
plt.grid(True)
plt.title('Количество заказов на покупателя')
plt.ylabel('Количество заказов');
```

Присутствуют выбросы по количеству заказов. Есть покупатель, сделавший 126 заказов.

```python
# Посмотрим 95-й и 99-й перцентили по количеству заказов
np.percentile(orders_by_customer['order_id'], [95, 99])
```

Не более 5 покупателей сделали более 2 заказов, и не более 1 процента покупателей сделали более 3 заказов.

<div class="alert alert-info" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid " ><b> Комментарий студента : </b>

Видим аномалию по диаграмме рассеяния - покупателя с большим количеством заказов (126).

</div>

```python
# Отфильтруем этого покупателя
abnormal_count = orders_by_customer[orders_by_customer['order_id'] == orders_by_customer['order_id'].max()]
orders = orders[~orders['customer_id'].isin(abnormal_count['customer_id'])]
```

## 2.2. Динамика выручки

```python
# Добавим в датафрейм столбец с месяцем
orders['month'] = orders['date'].astype('datetime64[M]')
orders['month']
```

```python
# Посчитаем выручку помесячно
revenue_by_month = orders.groupby(['month'])['revenue'].sum()
revenue_by_month
```

```python
# Построим график изменения общей выручки помесячно
revenue_by_month.plot(style='o-', grid=True)
plt.title('Динамика выручки')
plt.xlabel('Месяц')
plt.ylabel('Выручка');
```

Максимальная выручка была в декабре 2018 года. Минимальная выручка была в ноябре 2019 года.

## 2.3. Динамика среднего чека

```python
# Посчитаем средний чек заказа помесячно
revenue_by_order = orders.groupby(['month']).agg({'revenue':'sum','order_id':'nunique'}).reset_index()
revenue_by_order.columns = ['month', 'revenue_sum', 'order_count']
revenue_by_order['revenue_mean'] = revenue_by_order['revenue_sum'] / revenue_by_order['order_count']
revenue_by_order
```

```python
# Построим график изменения среднего чека помесячно
revenue_by_order.plot(x='month', y='revenue_mean', grid=True, style='o-')
plt.title('Динамика среднего чека')
plt.xlabel('Месяц')
plt.ylabel('Средний чек');
```

Максимальный средний чек был в ноябре 2018 года. Минимальный средний чек был в ноябре 2019 года.

## 2.4. Средняя выручка на покупателя

```python
# Посчитаем среднюю выручку на покупателя помесячно
revenue_by_customer = orders.groupby(['month']).agg({'revenue':'sum','customer_id':'nunique'}).reset_index()
revenue_by_customer.columns = ['month', 'revenue_sum', 'customer_count']
revenue_by_customer['revenue_mean'] = revenue_by_customer['revenue_sum'] / revenue_by_customer['customer_count']
revenue_by_customer
```

```python
# Построим график изменения средней выручки помесячно
revenue_by_customer.plot(x='month', y='revenue_mean', grid=True, style='o-')
plt.title('Динамика средней выручки')
plt.xlabel('Месяц')
plt.ylabel('Средняя выручка');
```

Максимальная средняя выручка была в ноябре 2018 года. Минимальная выручка была в ноябре 2019 года.

# 4. Сегментация покупателей по сумме покупок

## 4.1. Сегментация покупателей

```python
min_date = orders["date"].min()
print(f'Минимальная дата в датасете: {min_date}')
max_date = orders["date"].max()
print(f'Максимальная дата в датасете: {max_date}')
```

```python
# Cоздадим параметры recency, frequency и monetary по клиенту
last_date = orders['date'].max() + pd.Timedelta(days=1)
rfm = orders.groupby('customer_id',as_index=False).agg({'date': lambda x: (last_date - x.max()).days,
                                         'order_id': lambda x: x.nunique(), 
                                         'revenue': lambda x: x.sum()})
rfm.rename(columns={'date': 'recency', 
                    'order_id': 'frequency', 
                    'revenue': 'monetary'}, inplace=True)
rfm
```

```python
# Выставим оценки в диапазоне от 1 до 3
quantiles = rfm.quantile(q=[0.33, 0.66])

def RScore(x,p,d):
     if x <= d[p][0.33]:
         return 1
     elif x <= d[p][0.66]:
         return 2
     else:
         return 3

rfm['r_quartile'] = rfm['recency'].apply(RScore, args=('recency',quantiles,))
rfm['f_quartile'] = rfm['frequency'].apply(RScore, args=('frequency',quantiles,))
rfm['m_quartile'] = rfm['monetary'].apply(RScore, args=('monetary',quantiles,))
rfm['score'] = rfm.r_quartile.map(str) + rfm.f_quartile.map(str) + rfm.m_quartile.map(str)
rfm.head(10)
```

Используем сегментацию по сумме покупок:
1 - мелкие покупки,
2 - средние покупки,
3 - крупные покупки.

# 5. Формулирование и проверка гипотез

## 5.1. Проверка гипотез о различии в среднем чеке между сегментами покупателей

```python
# Напишем функцию для расчета статистической значимости с помощью критерия Манна-Уитни
def MW(df,cat1,cat2,param):
    results = stats.mannwhitneyu(rfm[rfm['score'].str[using_param] == cat1][param], 
                                 rfm[rfm['score'].str[using_param] == cat2][param])
    print('p-значение: ', '{0:.5f}'.format(results.pvalue))
    
    alpha = 0.05
    m = 3 # количество тестов
    
    # считаем поправку Бонферрони
    alpha_bonf = alpha / m
    
    if results.pvalue < alpha_bonf:
        print('Отвергаем нулевую гипотезу: разница статистически значима')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
```

```python
# Посчитаем средний чек для каждого покупателя
rfm['avg_order'] = rfm['monetary'] / rfm['frequency']
rfm
```

H0: Средний чек одинаковый для клиентов, которые совершали мелкие и средние покупки

H1: Средний чек различается для клиентов, которые совершали мелкие и средние покупки

```python
MW(rfm,'1','2','avg_order')
```

H0: Средний чек одинаковый для клиентов, которые совершали мелкие и крупные покупки

H1: Средний чек различается для клиентов, которые совершали мелкие и крупные покупки

```python
MW(rfm,'1','3','avg_order')
```

H0: Средний чек одинаковый для клиентов, которые совершали средние и крупные покупки

H1: Средний чек различается для клиентов, которые совершали средние и крупные покупки

```python
MW(rfm,'2','3','avg_order')
```

mannwhitneyu - это непараметрический тест, главное отличие которых состоит в том, что они работают с рангами, — номерами значений в упорядоченном ряду, — никак не учитывая сами значения. Поэтому к ним прибегают тогда, когда работа с самими значениями невозможна из-за выбросов, сильно сдвигающих параметрические результаты.
